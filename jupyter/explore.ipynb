```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
game_rating = {
	"Game": ["Dark Souls", "Dark Souls 2", "Dark Souls 3"],
	"Rating": [4, 4.3, 4.9]
}

df = pd.DataFrame(game_rating)
df
df.describe()
```

```python
plt.bar(df["Game"], df["Rating"], color = 'blue')
plt.title("FromSoftware Game Ratings", color = "")
plt.xlabel("Game titles", color = "red")
plt.ylabel("Rating", color = "red")
plt.show()
```

```python
sales = {
	"Year": [2022, 2023, 2024, 2025, 2026, 2027],
	"Sales": [200, 300, 400, 500, 600, 700]
}

df = pd.DataFrame(sales)
df
```

```python
plt.bar(df["Year"], df["Sales"])
plt.title("Car sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
pie = np.array(df["Sales"])
max_sales = np.argmax(pie) # returns the index of the max sales
```

```python
distinct_largest = True
for i in range(len(pie)):
	if (pie[i] == pie[max_sales] and i != max_sales):
		distinct_largest = False
		break

if (distinct_largest == True):
	highlight = [0.2 if sale == max_sales else 0 for sale in range(len(pie))]
else:
	highlight = [0 for sale in range(len(pie))]
print(highlight)
```

```python
plt.pie(pie, labels = df["Year"], explode = highlight)
plt.legend(title = "Car sales")
```
